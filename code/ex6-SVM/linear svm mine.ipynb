# 1. 线性SVM

```python
import numpy as np
import pandas as pd
import sklearn.svm
import seaborn as sns
import scipy.io as sio
import matplotlib.pyplot as plt
```

```python
mat = sio.loadmat('./data/ex6data1.mat')
data = pd.DataFrame(mat['X'],columns=['x1' ,'x2'])
data['y'] = mat['y']
print(data.head(5), data.shape)
```

## 数据可视化

```python
positive = data[data['y'].isin([1])]
negative = data[data['y'].isin([0])]
fig, ax = plt.subplots(figsize=(8,6))
ax.scatter(positive['x1'], positive['x2'], s=20, c='black', marker='x', label='positive')
ax.scatter(negative['x1'], negative['x2'], s=20, c='y', marker='o', label='newgative')
ax.legend()
ax.set_xlabel('x 1')
ax.set_ylabel('x 2')
plt.show()
```

```python
data_normalize = pd.DataFrame()
data_normalize['x1'] = (data['x1'] - data['x1'].mean()) / data['x1'].std()
data_normalize['x2'] = (data['x2'] - data['x2'].mean()) / data['x2'].std()
data_normalize['y'] = data['y']
data_normalize.head()
```

```python
svc1 = sklearn.svm.LinearSVC(C=1, loss='hinge')
svc1.fit(data_normalize[['x1', 'x2']], data_normalize['y'])
svc1.score(data_normalize[['x1', 'x2']], data_normalize['y'])
```

```python
data_normalize['SVM1 Confidence'] = svc1.decision_function(data_normalize[['x1', 'x2']])
#data_normalize['SVM1 Confidence']
```

```python
fig, ax = plt.subplots(figsize=(8,6))
ax.scatter(data_normalize['x1'], data_normalize['x2'], s=20, c=data_normalize['SVM1 Confidence'])
ax.set_xlabel('x 1')
ax.set_ylabel('x 2')
plt.show()
```

```python
svc2 = sklearn.svm.LinearSVC(C=100, loss='hinge',max_iter=2000)
svc2.fit(data_normalize[['x1', 'x2']], data_normalize['y'])
svc2.score(data_normalize[['x1', 'x2']], data_normalize['y'])
```

```python
data_normalize['SVM2 Confidence'] = svc2.decision_function(data_normalize[['x1', 'x2']])
```
